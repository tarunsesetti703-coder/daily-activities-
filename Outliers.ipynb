# Outliers

```python
import pandas as pd
```

```python
data=pd.DataFrame({
    "Study_Hours":[1,2,3,4,5,6,7,8,9]
})
data
```

```python
data.describe()
```

# IQR Method

```python
Q1=data.quantile(0.25)
Q3=data.quantile(0.75)

IQR=Q3-Q1

LB=Q1-1.5*IQR
UB=Q3+1.5*IQR

outliers=data[(data<(LB)) |(data>(UB))]
```

```python
outliers
```

# Z-score

#### Z-score methods

```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.DataFrame({
    "Study_Hours":[1,2,3,4,5,6,7,8,20]
})
data
```

```python
z_scores=np.abs(stats.zscore(data["Study_Hours"]))
outliers=data[z_scores > 2]
outliers
```

```python
plt.scatter(range(len(data)),data["Study_Hours"])
plt.xlabel("Index")
plt.xlabel("Study Hours")
plt.title("Scatter plot showing Outliers")
plt.show()
```

```python
plt.hist(data["Study_Hours"],40)
plt.xlabel("Study Hours")
plt.ylabel("Frequency")
plt.title("Histogram showing Outlier Effect")
plt.show()
```
