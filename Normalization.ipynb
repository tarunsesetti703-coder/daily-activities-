# Normalization

```python
import pandas as pd
```

```python
s=pd.Series([1000,2000,3000,5000,9000]) # min(1000)  max(9000)
s
```

```python
s.min()
```

```python
s.max()
```

```python
s_norm=(s-s.min())/(s.max()-s.min())
print(s_norm)
```
