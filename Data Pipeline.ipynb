# Data Pipeline

### Data pipeline:
 - A data pipeline is a step-by-step process that takes raw data and converts it into model-ready data.

#### Basic Flow:
 - Raw Data -> Cleaning -> Processing -> Model

#### Why do we need a Data Pipeline:
 - Real-World/Raw data is not clean:
       * Missing values
       * Duplicate records
       * Different formats
       * Noise and outliers

#### A data pipeline:
  * Organizes data properly
  * automates human errors
  * Reduces human errors
  * Makes ML models reliable

#### Pipeline in sklearn:
 - A Pipeline in sklearn is a sequence of data preprocessing steps and a machine learning model combined together to automate training and prediction.

- Data Collection --> Data Preparation --> ML Model Training --> ML Model Evalution --> ML Model Deployment

### Simple Data Pipeline Flow:
#### 1.Data Collection:
  - Collecting data from different sources.

##### Comman Sources:
  * CSV/Excel files
  * APIs

##### Example:
 - Student marks data collected from a CSV file.

#### 2.Data clening:
 - Fixing errors and problems in data.

##### Common Tasks:
 * Handling missing values
 * Removing duplicates
 * Fixing incorrect values
 * Removing outliers

##### Example:
 - Fill missing study hours with average
 - Remove duplicate student records

#### 3.Data preprocessing:
 - Converting data into a machine-understandable format.

##### Common Techniques:
 * Scaling (StandarScaler, MinMaxScaler)
 * Encoding (Label Encoding, One-Hot Encoding)
 * Train-Test-Split

##### Example:
 - Convert "Male/Female" -> 0/1
 - Scale marks to same range

#### 4.Feature Engineering:
 - Creating new and better features from existing data.

##### Common Methods:
 * Binning
 * Feature Creation
 * Feature Selection
 * Feature Transformation

##### Example:
 - Convert age -> Young/Adult/Senior
 - Create 'Total_Marks' from subject marks

#### 5.Model Training:
 - Giving processed data to a machine learning algorithm so it can learn patterns.

##### Common Algorithms:
 * Linear Regression
 * Logistic Regression
 * Decision Tree
 * Random Forest

#### 6.Model Evaluation:
 - Checking how well the model is performing.

 * Accuracy
 * Mean Squared Error (MSE)
 * Mean Absolute Error (MAE)
 * RÂ² score

##### Example:
 - Compare predicted marks with actual marks.
