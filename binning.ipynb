```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.DataFrame({
    "age":[18,22,25,30,35,40,45,50]
})
data
```

# Equal Width

```python
data["age_bin"]= pd.cut(data["age"],bins=4)
data
```

```python
plt.hist(data["age"],bins=4)
plt.xlabel("age")
plt.ylabel("Frequency")
plt.title("Equal-Width binning using pd.cut()")
plt.show()
```

# Equal frequency

```python
data=pd.DataFrame({
    "age":[18,22,25,30,35,40,45,50]
})
data
```

```python
data["age_bin"]= pd.qcut(data["age"],q=2)
data
```

```python
plt.hist(data["age"],bins=2)
plt.xlabel("age")
plt.ylabel("Frequency")
plt.title(" Equal Frequency binning using pd.qcut()")
plt.show()
```
