# Concate Data frames

```python
india_weather=pd.DataFrame({
    "city":["mumbai","banglore","hyderabad"],
    "Temperature":[32,45,35]
})
india_weather
```

```python
US_weather=pd.DataFrame({
    "city":["New york","Chicago","orlando"],
    "Temperature":[32,45,35]
})
US_weather
```

```python
df=pd.concat([india_weather,US_weather])
df
```

```python
df=pd.concat([india_weather,US_weather],axis=1)
df
```

# Merge DataFrames

```python
temp_df=pd.DataFrame({
    "city":["Banglore","Hyderabad","vijayawada"],
    "Temperature":[49,47,17]
})
temp_df
```

```python
humidity_df=pd.DataFrame({
    "city":["Kerela","Banglore","Hyderabad",],
    "Humidity":[17,7,30]
})
humidity_df
```

```python
df=pd.merge(temp_df,humidity_df,on="city",how="outer")
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Temperature']=df['Temperature'].fillna(df.Temperature.min())
```

```python
df.isnull()
```

```python
df
```

```python
df['Humidity']=df['Humidity'].fillna(df.Humidity.min())
df.isnull()
```
