```python
b=np.ones(4)+1
print("a=",a)
print("b=",b)
```

```python
print(a*b)
```

```python
print(a+b)
```

# Comparision Operators

```python
a=np.array([1,2,3,4])
b=np.array([5,2,3,4])
print(a==b)
```

```python
print(a>b)
```

```python
a=np.array([1,2,3,4])
b=np.array([5,2,3,4])
c=np.array([1,2,3,4])
print(np.array_equal(a,b))
```

```python
print(np.array_equal(a,c))
```

# Logical operators

```python
a=np.array([1,0,1,0],dtype="bool")
b=np.array([1,1,0,0],dtype="bool")
```

```python
print(np.logical_or(a,b))
```

```python
print(np.logical_and(a,b))
```
